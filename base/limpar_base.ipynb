```python
#Importando bibliotecas
import pandas as pd
import numpy as np
```

```python
#Lendo CSV
df = pd.read_csv('3 - MetalForte_300_Registros.csv')
```

```python
#Arrumando as datas
df['Data_Manutencao'] = pd.to_datetime(
    df['Data_Manutencao'],
    dayfirst=True,
    errors='coerce',
    format='mixed')
```

```python
#Normalizando valores da coluna Máquina
df['Maquina'] = np.where(df['Maquina'] == 'Torno C N C', 'Torno CNC', df['Maquina'])
df['Maquina'] = np.where(df['Maquina'] == 'TornoCNC', 'Torno CNC', df['Maquina'])
df['Maquina'] = np.where(df['Maquina'] == 'Prensa Hidraulica', 'Prensa Hidráulica', df['Maquina'])
```

```python
#Corrigindo valores em setor
df['Setor'] = np.where(df['Setor'] == 'Usin.', 'Usinagem', df['Setor'])
```

```python
#Corrigindo os valores para positivo
df['Custo_Manutencao'] = df['Custo_Manutencao'] * -1
```

```python
#Removendo colunas desnecessárias
df = df.drop(columns=['Hora_Inicio', 'Hora_Fim', 'Observacao', 'Tecnico', 'Especialidade'])
```

```python
#Corrigindo a coluna Horas_Parada
df['Horas_Parada'] = df['Horas_Parada'] * -1

df['Horas_Parada'] = pd.to_timedelta(df['Horas_Parada'], unit='h')

df['Horas_Parada'] = df['Horas_Parada'].astype(str).str.replace('0 days ', '', regex=False)
```

```python
df.to_csv('Metalforte_limpa.csv', index=False)
```
