```python
#Importando bibliotecas
import pandas as pd
import numpy as np
import random as r
```

```python
#Lendo arquivos
df = pd.read_csv('3 - MetalForte_300_Registros.csv')
tec = pd.read_csv('tabela_tecnicos.csv', sep=';')
```

```python
#Removendo colunas desnecessárias
df.drop(['Hora_Inicio', 'Hora_Fim', 'Observacao', 'Tecnico'], axis=1, inplace=True)
```

```python
df['Data_Manutencao']
```

```python
#Normalizando as datas
df['Data_Manutencao'] = pd.to_datetime(df['Data_Manutencao'], format='mixed')

df['Data_Manutencao']
```

```python
#Normalizando o horário
df['Horas_Parada'] = df['Horas_Parada'] * -1

df['Horas_Parada'] = (
    pd.to_datetime('00:00:00')
    + pd.to_timedelta(df['Horas_Parada'], unit='h')
).dt.strftime('%H:%M')
```

```python
df['Maquina'].unique()
```

```python
#Normalizando valores
df['Maquina'] = np.where(df['Maquina'] == 'Torno C N C', 'Torno CNC', df['Maquina'])
df['Maquina'] = np.where(df['Maquina'] == 'TornoCNC', 'Torno CNC', df['Maquina'])
df['Maquina'] = np.where(df['Maquina'] == 'Prensa Hidraulica', 'Prensa Hidráulica', df['Maquina'])

df['Maquina'].unique()
```

```python
#Convertendo valores para positivo
df['Custo_Manutencao'] = df['Custo_Manutencao'] * -1
df['Custo_Manutencao']
```

```python
#Trazendo valores aleatórios da tabela técnicos
id = tec['ID_Tecnico'].to_list()

df['ID_Tecnico'] = np.random.choice(id, size=len(df))

df = df.merge(
    tec[['ID_Tecnico', 'Nome_Tecnico', 'Especialidade']],
    on='ID_Tecnico',
    how='left'
)
```

```python
df.to_excel('Metalforte_limpa.xlsx', index=False)
```
